```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

#1 Import the given “Salary_Data.csv”
data = pd.read_csv('Salary_Data.csv')
print(data.head())

#2 Split the data in train_test partitions, such that 1/3 of the data is reserved as test subset.
X = data[['YearsExperience']]
y = data['Salary']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

#3 Train and predict the model.
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

#4 Calculate the mean_squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

#5 Visualize both train and test data using scatter plot
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.scatter(X_test, y_test, color='green', label='Test data')
plt.plot(X_train, model.predict(X_train), color='red', linewidth=2, label='Regression line')

plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression')
plt.legend()

plt.show()

```
